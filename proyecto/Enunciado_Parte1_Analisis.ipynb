4- Una métrica de evaluación relevante para el negocio es maximizar los tickets respondidos al mes. Otra métrica es optimizar la cantidad de tickes criticos solucionados y tiempo de respuesta. (FALTA JUSTIFICAR)

5- La interpretación de las métricas respecto al problema binario de clasificación es: 
- Exactitud (Accuracy): Mide el total de predicciones correctas divididas por el total de datos. Es la métrica más rápida, pero puede ser engañosa si algunas clases tienen muchos más datos que otras.
- Precisión (Precision): La proporción de predicciones correctas para una clase específica respecto a todas las veces que el modelo predijo esa clase.
- Exhaustividad (Recall): La proporción de predicciones correctas para una clase respecto al total real de elementos que pertenecen a esa clase.
- Puntuación F₁ (F1-Score): Combina la precisión y la exhaustividad en una sola medida. Es muy útil si necesitas un equilibrio entre ambas métricas
Por el lado de regresión seria:
- R^2
- RMSE
- MAE (FALTA)
- MSE

Finalmente, para este problema la métrica más importante es el Recall, ya que buscamos minizar los falsos negativos (que clasifique un ticket con categoría baja cuando debería ser crítico). Adicionalmente, nuestra segunda métrica clave también es el F1-Score, ya que busca encontrar el equilibrio entre falsos positivos y falsos negativos.

6- Para las clases Crítico y Alto, se prioriza la métrica Recall, ya que buscamos maximizar la captura de casos graves. Para las clases media y baja sería precisión, lo cual mide la certeza de las alertas generadas, buscando asegurar que la anomalías sean precisas en estas clases.

7- La agregación macro es una métrica global que calcula el promedio simple de las métricas de cada clase, tratando a todos por igual. Mientras que, la agregación weighted calcula la métrica para cada clase y la multiplica por la proporción de instancias (tickets) que pertenencen a cada prioridad, generando un peso que refleja el desbalance de los datos. La agregación que utilizaríamos sería la macro, ya que buscamos un alta sensibilidad en la clase crítica, castigando al modelo en caso de fallar en una clase, dando mismo peso en promedio a todas las clases. De lo contrario, una agregación Weighted puede sesgar los resultados al presentar un desbalance entre clases.

```python
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots

# 1. Definimos explícitamente cuáles variables queremos analizar
# Excluimos IDs, fechas y texto libre para que los histogramas tengan sentido
columnas_numericas = [
    "N_Caracteres_Ticket",
    "Usuario-Antiguedad_Cuenta_Dias",
    "Tiempo_de_Resolución_Horas",
]

columnas_categoricas = [
    "Canal_Ticket",
    "Categoría_Problema",
    "Usuario-Tipo_de_Cuenta",
    "Nivel_Prioridad",
]

columnas_totales = columnas_numericas + columnas_categoricas

# 2. Calculamos las dimensiones de la grilla de forma dinámica
total_variables = len(columnas_totales)
columnas_grilla = 2
filas_grilla = int(np.ceil(total_variables / columnas_grilla))

# 3. Creamos la grilla con los títulos correspondientes
fig_subplots = make_subplots(
    rows=filas_grilla,
    cols=columnas_grilla,
    subplot_titles=[f"Dist. de {col}" for col in columnas_totales],
)

# 4. Iteramos usando enumerate para no dejar ninguna columna en el camino
for idx, col in enumerate(columnas_totales):
    fila = (idx // columnas_grilla) + 1
    columna = (idx % columnas_grilla) + 1

    # Personalizamos el orden si es la columna de Nivel_Prioridad
    if col == "Nivel_Prioridad":
        fig_temporal = px.histogram(
            df, x=col, category_orders={col: ["Baja", "Media", "Alto", "Crítico"]}
        )
    elif col == "Tiempo_de_Resolución_Horas":
        # Usamos log_y=True para que las frecuencias bajas no desaparezcan
        fig_temporal = px.histogram(df, x=col, log_y=True)
    else:
        fig_temporal = px.histogram(df, x=col)

    # Extraemos la traza con fig.data[0] tal como lo pide el Hint
    fig_subplots.add_trace(fig_temporal.data[0], row=fila, col=columna)

# 5. Desvinculamos los ejes para que cada variable mantenga su escala e independencia
fig_subplots.update_xaxes(matches=None)
fig_subplots.update_yaxes(matches=None)

# 6. Ajustamos el tamaño total para que las 4 filas de gráficos se vean perfectas
fig_subplots.update_layout(
    height=filas_grilla * 320,
    width=950,
    title_text="<b>Análisis Exploratorio: Distribución de Características de Tickets</b>",
    showlegend=False,
)

fig_subplots.show()
```

```python
variables_numericas = [
    "N_Caracteres_Ticket",
    "Usuario-Antiguedad_Cuenta_Dias",
    "Tiempo_de_Resolución_Horas",
]
df_numerico = df[variables_numericas]

matriz_corr = df_numerico.corr()

# 3. Graficamos el mapa de calor utilizando imshow
fig_corr = px.imshow(
    matriz_corr,
    text_auto=".2f",  # Muestra los valores con 2 decimales dentro de cada cuadro
    aspect="auto",
    color_continuous_scale="RdBu_r",  # Escala de color clásica (Rojo = Positiva, Azul = Negativa)
    labels=dict(color="Correlación"),
    x=matriz_corr.columns,
    y=matriz_corr.columns,
    title="<b>Matriz de Correlación entre Variables Numéricas</b>",
)

# Ajustamos tamaño para que sea perfectamente legible
fig_corr.update_layout(width=600, height=500)

fig_corr.show()
```

```python

etiquetas = ["Bajo", "Medio", "Alto"]


df["N_Caracteres_Ticket_Bin"] = pd.qcut(
    df["N_Caracteres_Ticket"], q=3, labels=etiquetas, duplicates="drop"
)
df["Tiempo_de_Resolución_Horas_Bin"] = pd.qcut(
    df["Tiempo_de_Resolución_Horas"], q=3, labels=etiquetas, duplicates="drop"
)
df["Usuario-Antiguedad_Cuenta_Dias_Bin"] = pd.qcut(
    df["Usuario-Antiguedad_Cuenta_Dias"], q=3, labels=etiquetas, duplicates="drop"
)


variables_analisis = [
    "Categoría_Problema",
    "Canal_Ticket",
    "Usuario-Tipo_de_Cuenta",
    "N_Caracteres_Ticket_Bin",
    "Tiempo_de_Resolución_Horas_Bin",
    "Usuario-Antiguedad_Cuenta_Dias_Bin",
]

orden_prioridades = ["Baja", "Media", "Alto", "Crítico"]


fig_subplots = make_subplots(
    rows=3,
    cols=2,
    subplot_titles=[f"Contingencia: {var} vs Prioridad" for var in variables_analisis],
)


for idx, var in enumerate(variables_analisis):
    fila = (idx // 2) + 1
    columna = (idx % 2) + 1

    tabla_contingencia = pd.crosstab(
        df[var], df["Nivel_Prioridad"], normalize="index"
    )
    columns_existentes = [
        p for p in orden_prioridades if p in tabla_contingencia.columns
    ]
    tabla_contingencia = tabla_contingencia[columns_existentes]

    fig_temporal = px.imshow(
        tabla_contingencia,
        text_auto=".2f",
        color_continuous_scale="YlOrRd",
    )


    fig_subplots.add_trace(fig_temporal.data[0], row=fila, col=columna)


    fig_subplots.update_xaxes(
        tickmode="array",
        tickvals=list(range(len(columns_existentes))),
        ticktext=columns_existentes,
        row=fila,
        col=columna,
    )
    fig_subplots.update_yaxes(
        tickmode="array",
        tickvals=list(range(len(tabla_contingencia.index))),
        ticktext=[str(i) for i in tabla_contingencia.index],
        row=fila,
        col=columna,
    )


fig_subplots.update_layout(
    height=1000,
    width=950,
    title_text="<b>Análisis de Contingencias Normalizadas por Fila (vs Nivel_Prioridad)</b>",
    coloraxis={"colorscale": "YlOrRd"},
)

fig_subplots.show()
```

```python
df['Fecha_Envío'] = df['Fecha_Envío'].astype('datetime64[ns]')
df['Fecha_Envio_Mes'] = df['Fecha_Envío'].dt.month

df_conteo_mensual = (
    df.groupby(['Fecha_Envio_Mes', 'Nivel_Prioridad']).size().reset_index(name='Cantidad_reclamos_mes')
)


orden_prioridades = ["Baja", "Media", "Alta", "Critica"]

fig_lineas = px.line(
    df_conteo_mensual,
    x="Fecha_Envio_Mes",
    y="Cantidad_reclamos_mes",
    color="Nivel_Prioridad",  # Crea una serie/línea por cada nivel de prioridad
    category_orders={"Nivel_Prioridad": orden_prioridades},
    title="<b>Evolución Mensual de Reclamos por Nivel de Prioridad</b>",
    labels={
        "Fecha_Envio_Mes": "Mes de Envío",
        "Cantidad_reclamos_mes": "Cantidad de Reclamos",
        "Nivel_Prioridad": "Prioridad",
    },
    markers=True,  # Agrega puntitos en cada mes para que sea más fácil leer los valores precisos
)

# 4. Ajustes de diseño
fig_lineas.update_layout(
    width=900,
    height=500,
    xaxis_tickangle=-45,  # Inclina los meses para que no se encimen en el eje X
)

fig_lineas.show()
```
