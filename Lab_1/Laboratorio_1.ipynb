### Extra: ¿Qué es el CI y qué verifican estas herramientas?

**CI (Integración Continua)** es una práctica en la que cada vez que se sube código al repositorio (mediante un push o un Pull Request), se ejecutan automáticamente una serie de verificaciones para asegurar que el código cumple ciertos estándares de calidad. En este curso, el CI está configurado en `.github/workflows/ci.yml` y es ejecutado por GitHub Actions.

Las herramientas que configuramos cumplen roles complementarios:

| Herramienta | ¿Qué verifica? |
|---|---|
| **`ruff` (linter)** | Detecta errores de estilo, imports no usados, variables indefinidas, patrones propensos a bugs y sintaxis desactualizada |
| **`ruff` (formatter)** | Formatea el código automáticamente para que tenga un estilo consistente (espacios, saltos de línea, comillas, etc.) |
| **`pre-commit`** | Ejecuta un conjunto de revisiones automáticamente *antes de cada commit*, impidiendo subir código que no pase las verificaciones |

En la práctica: cuando hagan `git commit`, `pre-commit` correrá `ruff` sobre sus archivos. Si hay problemas, el commit será bloqueado y deberán corregirlos antes de continuar. Esto asegura que todo el código que llegue al repositorio sea consistente y libre de errores básicos.

#### Pregunta 1 (0.5 pts)

Dada una lista de $n$ elementos, no necesariamente ordenada, defina funciones que busquen:
- El elemento menor.
- El elemento mayor.
- El promedio de los elementos en la lista.

Ejemplo:
```python
lista_n = [12,4,34,1,2,52,1,54,98,5,235,76,57,87,123,32,32]
menor = 1
mayor = 235
```


````python
############ Su Código Aquí####################
listaN = [12, 4, 34, 1, 2, 52, 1, 54, 98, 5, 235, 76, 57, 87, 123, 32, 32]


def mayor_elemento(lista):
    if len(lista) == 0:  # Manejar el caso de lista vacía
        return None
    mayor = lista[0]
    for valor in lista:
        if valor > mayor:
            mayor = valor
    return mayor


def menor_elemento(lista):
    if len(lista) == 0:  # Manejar el caso de lista vacía
        return None
    menor = lista[0]  # Usar valores de la lista en lugar de valores predeterminados
    for valor in lista:
        if valor < menor:
            menor = valor
    return menor


def promedio_elemento(lista):
    if len(lista) == 0:  # Manejar el caso de lista vacía
        return None
    suma = 0
    for valor in lista:
        suma = suma + valor
    promedio = suma / len(lista)
    return promedio


menor, mayor, promedio = menor_elemento(listaN), mayor_elemento(listaN), promedio_elemento(listaN)

print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)


# Casos borde: (Lista con numeros negativos, lista con numeros mayores a 999, lista con numeros repetidos y lista vacia)

print("\nTest con lista de numeros negativos: \n")
Negative_list = [-5, -15, -42, -3, -100, -7]

menor, mayor, promedio = menor_elemento(Negative_list), mayor_elemento(Negative_list), promedio_elemento(Negative_list)
print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)

print("\nTest con lista de numeros mayores a 999: \n")
Big_Numbers_list = [1000, 1500, 2048, 9999, 54321, 1001]

menor, mayor, promedio = (
    menor_elemento(Big_Numbers_list),
    mayor_elemento(Big_Numbers_list),
    promedio_elemento(Big_Numbers_list),
)
print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)

print("\nTest con lista de numeros repetidos \n")
Repeated_Numbers_list = [7, 42, 7, 15, 99, 42, 7, 10]
menor, mayor, promedio = (
    menor_elemento(Repeated_Numbers_list),
    mayor_elemento(Repeated_Numbers_list),
    promedio_elemento(Repeated_Numbers_list),
)
print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)

print("\nTest con lista vacia \n")

Empty_list = []
menor, mayor, promedio = menor_elemento(Empty_list), mayor_elemento(Empty_list), promedio_elemento(Empty_list)
print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)

print("\nTest con lista con un solo valor \n")
list_unique = [1]
menor, mayor, promedio = menor_elemento(list_unique), mayor_elemento(list_unique), promedio_elemento(list_unique)
print("El valor menor es: ", menor)
print("El valor mayor es: ", mayor)
print("El promedio de la lista es: ", promedio)
````

#### Pregunta 2 (0.5 pts)
Genere una función que obtenga la moda de una lista usando diccionarios.

````python
############ Su Código Aquí####################
def buscar_moda(lista):
    if len(lista) == 0:  # Manejar el caso de lista vacía
        return None
    frecuencia = {}
    for valor in lista:
        if valor in frecuencia:
            frecuencia[valor] += 1
        else:
            frecuencia[valor] = 1
    valor_moda = 0
    frecuencia_moda = 0
    for valor in lista:
        if frecuencia[valor] > frecuencia_moda:
            valor_moda = valor
            frecuencia_moda = frecuencia[valor]
    return valor_moda, frecuencia_moda #Devuelve primera moda
````

````python
#Test

# Casos normales
assert buscar_moda([7, 42, 7, 15, 99, 42, 7, 10]) == (7,3), "Error en moda normal"

# Lista vacía
assert buscar_moda([]) is None,  "Error en moda vacia"

# Un elemento
assert buscar_moda([1]) == (1,1), "Error en moda de un elemento"

# Todos distintos
assert buscar_moda([4,8,15,16,23,42]) == (4,1), "Error en moda de todos distintos"

# Todos iguales
assert buscar_moda([5,5,5,5,5]) == (5,5), "Error en moda de todos iguales"

# Moda negativa
assert buscar_moda([-3,-3,-1,-2,-3]) == (-3,3), "Error en moda negativa"

# Empate en frecuencia
assert buscar_moda([1,1,2,2,3]) == (1,2), "Error en moda de iguales frecuencia"

print("Todos los tests pasaron correctamente")
````

#### Pregunta 3 (0.5 pts)
Vea si un número es palíndromo o no. Los números palíndromos son aquellos que al leerlos al revés y al derecho son iguales, por ejemplo el 123321 o el 98789.

````python
############ Su Código Aquí####################
def is_palindrome(numero):
    # Verficar si el input es un int
    if not isinstance(numero, int):
        raise ValueError("El número debe ser un entero.")
    # Manejar el caso de número negativo
    if numero < 0:
        numero = -numero
    numero_str = str(numero)
    palindromo = numero_str == numero_str[::-1]
    if palindromo:
        print("Es un palindromo")
    else:
        print("No es palindromo")
    return palindromo


# Test cases para IsPalindrome

# 1) Cero (caso borde mínimo no negativo)
assert is_palindrome(0) is True

# 2) Número negativo (la función lo normaliza a positivo)
assert is_palindrome(-12321) is True

# 3) Número con dígitos repetidos (caso borde de palíndromo con caracteres repetidos)
assert is_palindrome(1221) is True

# 4) Número con dígitos únicos (caso borde de numeros de digitos únicos)
assert is_palindrome(3) is True

# 5) Número con dígitos no palíndromos (caso borde de número no palíndromo)
assert is_palindrome(123) is False

# 6) Algo que no es un número (caso borde de tipo de dato incorrecto)
try:
    is_palindrome("abc")
except ValueError as e:
    print(e)
````

#### Pregunta 4 (0.5 pts)
Dado un texto, conviértalo en estilo Camel Case, vea el siguiente ejemplo:

    Texto inicial: “Este es un texto de ejemplo para mostrar el Camel Case”.

    Texto final: “EsteEsUnTextoDeEjemploParaMostrarElCamelCase”.

````python
############ Su Código Aquí####################
texto_test = "Este es un texto de ejemplo para mostrar el Camel Case"


# SnakeSpace
def camel_case(texto):
    if not isinstance(texto, str):
        raise ValueError("El texto debe ser un string.")
    texto_dividido = texto.split()
    texto_camelCase = ""
    for palabra in texto_dividido:
        palabra = palabra.capitalize()  # Convertir la primera letra a mayúscula y el resto a minúscula
        texto_camelCase += palabra
    return texto_camelCase


# Test cases para CamelCase
assert camel_case(texto_test) == "EsteEsUnTextoDeEjemploParaMostrarElCamelCase"
assert camel_case("") == ""
assert camel_case("a") == "A"
assert camel_case("hello world") == "HelloWorld"
assert camel_case("hola   mundo") == "HolaMundo"
assert camel_case("camel case") == "CamelCase"
try:
    camel_case(123)
except ValueError as e:
    print(e)
````

#### Pregunta 1 (0.5 pts):

Un monto inicial de dinero (por ejemplo un millón de pesos) se puede depositar en un banco con un interés anual de 5%. Esto significa  que  cumplido  un  año  se  dispondrá  de  \\$1.050.000. Si se deposita por dos años, entonces  al  finalizar  el  2º  año se dispondrá de \\$1.102.500,  es  decir,  \\$1.050.000  reajustado  en  5%.  

Al  respecto, si  se  invoca  a  la  función  `calc_capital_final(1000000, 5, 2)`  entrega  el resultado `1050000.0`. (Posible error en el enunciado, el resultado correcto debería ser `1102500.0`)

Escriba la función `calc_capital_final`

$$C_f = C_i(1+i)^n$$

- $C_f$ = Capital Final
- $C_i$ = Capital Inicial
- $i$ = Tasa de interés
- $n$ = Periodo del ahorro

Además, incluya tests que certifiquen el correcto funcionamiento de la función.

````python
def calc_capital_final(capital_inicial, tasa_interes, años):
    interes_decimal = tasa_interes / 100  # Convertir la tasa de interés a decimal
    interes = (1 + interes_decimal) ** años
    capital_final = capital_inicial * interes
    return capital_final


# Test cases para Calcular Capital Final
assert calc_capital_final(1000000, 5, 2) == 1102500.0
assert calc_capital_final(500000, 3, 5) == 579637.03715
assert calc_capital_final(0, 5, 10) == 0.0
assert calc_capital_final(1000000, 0, 10) == 1000000.0
assert calc_capital_final(1000000, 5, 0) == 1000000.0
````

#### Pregunta 2 (0.5 pts):

Un anagrama es una palabra que resulta de la transposición de letras de otra palabra o frase.

Por ejemplo, las palabras “caso” y “caos” sonan agramas de “saco”  y las palabras “acaso” y “ocaso” no son anagramas de “caso”. Las anagramas poseen las mismas letras, con la misma cantidad de apariciones, pero en un orden diferente

Escriba la función `def anagrama(x,y)` que reciba dos palabras y entregue `True` si son anagramas y `False` si no.


````python
############ Su Código Aquí####################
def anagram(x, y):
    if not isinstance(x, str) or not isinstance(y, str):
        raise ValueError("Ambos inputs deben ser strings.")
    x = x.replace(" ", "").lower()  # Eliminar espacios y convertir a minúsculas
    y = y.replace(" ", "").lower()  # Eliminar espacios y convertir a minúsculas
    return sorted(x) == sorted(y)


# Test cases para Anagram
assert anagram("listen", "silent") is True
assert anagram("triangle", "integral") is True
assert anagram("apple", "pabble") is False
assert anagram("Dormitory", "Dirty Room") is True
try:
    anagram(123, "silent")
except ValueError as e:
    print(e)
try:
    anagram("listen", 456)
except ValueError as e:
    print(e)
````
